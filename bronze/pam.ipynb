```python
import os
```

```python
rawzone = '/dbfs/mnt/usp/ProjetoIntegrador/rawzone/pam/'
bronze = '/dbfs/mnt/usp/ProjetoIntegrador/bronze/pam/'
```

```python
files_path = []
for root, _, files in os.walk(rawzone):
    for file in files:
        files_path.append(os.path.join(root,file))
```

```python
def snake_case(s):
    from re import sub
    from unidecode import unidecode
    return sub('[^a-z0-9_]', '', unidecode('_'.join(
        sub('([A-Z][a-z]+)', r' \1',
        sub('([A-Z]+)', r' \1',
        s.replace('-',' '))).split()).lower())).replace('__','_')
```

```python
import pandas as pd
import pyspark.sql.functions as F
import uuid

dfs = []

for file_path in files_path:
    year = file_path.split("/")[7]
    tipo_lavoura = file_path.split("/")[8]
    try:
        pdf = pd.read_excel(file_path, skiprows=4, engine='openpyxl')
        pdf = pdf.rename(columns={'Unnamed: 0':'regiao'})
        pdf['regiao'] = pdf['regiao'].apply(lambda x : x.strip())
        pdf.columns = [snake_case(column) for column in pdf.columns]
        pdf = pdf.replace('..', None).replace('...', None).replace('-', None)
        for column in pdf.columns[1:]:
            pdf[column] = pd.to_numeric(pdf[column])
        pdf['tipo_lavoura'] = snake_case(tipo_lavoura)
        pdf['file_path'] = file_path
        path = f'{bronze}/year={year}/'
        os.makedirs(path, exist_ok=True)
        pdf.to_parquet(f'{bronze}/year={year}/{uuid.uuid4().hex}.parquet')
    except Exception as e:
        print(file_path,e)
```
